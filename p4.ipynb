```python
import matplotlib.pyplot as plt
import numpy as np
plt.suptitle("GRAPH")
x=np.array([1,9])
y=np.array([2,9])
plt.subplot(4,2,1)
plt.ylabel("Blue color")
plt.xlabel("Blue color")
plt.plot(x,y,'o')

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])
x1=np.array([0,1,2,3])
y1=np.array([6,3,7,2])
plt.subplot(4,2,2)
plt.plot(x,y,color="orange")
plt.plot(x1,y1,linestyle="dotted",color="blue",linewidth=2.5)

x2=np.array([0,3,5,7])
y2=np.array([1,5,3,6])
plt.subplot(4,2,3)
plt.plot(x2,y2,marker='o',mfc='y',mec='b',ms=10)


x3=np.array([1,9])
y3=np.array([2,9])

plt.subplot(4,2,4)
plt.plot(x3,y3)
plt.grid()
plt.title('title2')



x4=np.array([2,5,6,5,1,3])
y4=np.array([3,2,5,5,2,3])

plt.subplot(4,2,5)
plt.scatter(x4,y4,color='r',alpha=0.5)
x4=np.array([4,7])
y4=np.array([4,6])
plt.subplot(4,2,5)
plt.scatter(x4,y4)

x6=np.array(["A","B","C"])
y6=np.array([15,45,65])
plt.subplot(4,2,6)
plt.bar(x6,y6,color='r',width=0.1)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35, 20, 20, 25])
mylabels = ["35", "20", "20", "25"]

myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True,startangle=90)
plt.legend()
plt.show()
```
